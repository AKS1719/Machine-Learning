```python
import pandas as pd 
```

```python
df = pd.read_csv('E:\\MachineLearning\\1LinearRegressionClass\\onehotEncodingMethod\\data.csv')
df
```

```python
dummies = pd.get_dummies(df.town,dtype=int)
```

```python
dummies 
```

```python
merged = pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
final = merged.drop(['town','west windsor'],axis='columns')
```

```python
final
```

```python
from sklearn.linear_model import LinearRegression

regressor  = LinearRegression()
```

```python
x = final.drop('price',axis='columns')
x
```

```python
y = final.price
```

```python
regressor.fit(x,y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
regressor.predict([[2800,0,1]])
```

```python
regressor.predict([[3400,0,0]])
```

```python
regressor.score(x,y)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
dfle = df
```

```python
dfle.town = labelencoder.fit_transform(dfle.town)
```

```python
dfle
```

```python
X = dfle[['town','area']].values
Y = dfle.price
```

```python
print(X,Y)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([('town',OneHotEncoder(),[0])],remainder='passthrough')


X = ct.fit_transform(X)
```

```python
X
```

```python
# [:,1:] the first one is for column scliceing and the next one is for column
X = X[:,1:]
X
```

```python
new_regressor = LinearRegression()
new_regressor.fit(X,Y)
```

```python
new_regressor.predict([[1,0,2800]])
```

```python
new_regressor.score(X,Y)
```
