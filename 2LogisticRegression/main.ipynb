```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("E:\\MachineLearning\\2LogisticRegression\\insuranceData.csv")
df.head()
```

```python
plt.scatter(df.age,df.bought_insurance)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[['age']],df.bought_insurance,train_size=0.9,test_size=0.1,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression()
```

```python
logistic_model.fit(x_train,y_train)
```

```python
print(x_test)
print(y_test)
logistic_model.predict(x_test)
```

```python
logistic_model.score(x_test,y_test)
```

```python
logistic_model.predict_log_proba(x_test)
```
