```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from word2number import w2n
```

```python
df = pd.read_csv('E:\\MachineLearning\\1LinearRegressionClass\\Excersise_multipleLinearRegression\\interview.csv')
df
```

```python
df = df.fillna({'experience': 0})

for i in range(len(df['experience'])):
    if df['experience'][i]!=0:
        print(df['experience'][i])
        df['experience'][i] = w2n.word_to_num(df['experience'][i]);
        
df
```

```python
import math 
median_test_score = math.floor(df.testScore.median())
median_test_score
```

```python
df = df.fillna({'testScore': median_test_score})
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(df[['experience','testScore','interviewScore']],df.salary)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
regressor.predict([[2,9,6]])
```

```python
y_pred = regressor.predict([[12,10,10]])
y_pred
```
